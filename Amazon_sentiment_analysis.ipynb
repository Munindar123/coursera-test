```

import os
import nltk
import nltk.corpus
import re
import matplotlib.pyplot as plt
```

```
text = open("/content/sample_data/Amzn_123", "r")
text = text.read().lower()
text = re.sub(r"[0-9:;\"'(),.$#%@-]", "", text)
```

```
 nltk.download('punkt')
```

```
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)

print(tokenized_word)
```

```
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
print(fdist)
print(fdist.most_common())
```

```
 nltk.download('stopwords')
```

```
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```
filtered_sentence = [w for w in tokenized_word if not w in stop_words] 
  
filtered_sentence = [] 
  
for w in tokenized_word: 
    if w not in stop_words: 
        filtered_sentence.append(w) 

  
 
print(filtered_sentence) 
```

```
positiveWordList = open("/content/sample_data/PositiveWords.txt","r")

positiveWordList = positiveWordList.read().lower()
print(positiveWordList)
```

```
numPosWords = 0

for word in tokenized_word:
  if word in positiveWordList:
            numPosWords  += 1
```

```
print(numPosWords)
```

```
negativeWordList = open("/content/sample_data/NegativeWords.txt","r")

negativeWordList = negativeWordList.read().lower()
```

```
numNegWords = 0

for word in tokenized_word:
        if word in negativeWordList:
            numNegWords  += 1
```

```
print(numNegWords)
```

```
 pol_score = (numPosWords- numNegWords) / ((numPosWords + numNegWords) + 0.000001)
 print(pol_score)
```

```
pos_score = numPosWords/len(filtered_sentence)*100
print(pos_score)
```

```
neg_score = numNegWords/len(filtered_sentence)*100
print(neg_score)
```
