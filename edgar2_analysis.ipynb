```
positiveWordList = open("/PositiveWords.txt","r")

positiveWordList = positiveWordList.read().lower()
print(positiveWordList)
```

```
numPosWords = 0

for word in filtered_sentence:
  if word in positiveWordList:
            numPosWords  += 1

print(numPosWords)
```

```
negativeWordList = open("/NegativeWords.txt","r")
```

```
negativeWordList = negativeWordList.read().lower()

numNegWords = 0

for word in filtered_sentence:
        if word in negativeWordList:
            numNegWords  += 1

print(numNegWords)
```

```
neutral_words = int(len(filtered_sentence))-(numNegWords+numPosWords)
```
