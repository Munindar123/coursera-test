```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```
df = pd.read_csv("Combined_News_DJIA.csv")
```

```
df.head()
```

```
df.shape
```

```
df.tail(400)
```

```
train = df[df['Date'] < '2014-12-01']
test = df[df['Date'] > '2014-11-30']
```

```
train.tail()
```

```
test.head()
```

```
data = train.iloc[:,2:]
```

```
data.head()
```

```
data.replace("[^a-zA-Z]" , " " ,regex=True, inplace=True)
```

```
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head()
```

```
for index in new_index:
  data[index] = data[index].str.lower()
```

```
data.head()
```

```
headlines = []
for row in range(0,len(data.index)):
  headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```
headlines[0]
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```
countvectors = CountVectorizer(ngram_range=(2,2))
traindataset = countvectors.fit_transform(headlines)
```

```
randomforest = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomforest.fit(traindataset,train['Label'])
```

```
test_transform = []
for row in range(0,len(test.index)):
  test_transform.append(' '.join(str(x) for x in test.iloc[row,2:]))

test_dataset = countvectors.transform(test_transform)
predictions = randomforest.predict(test_dataset)
```

```
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
matrix = confusion_matrix(test['Label'],predictions)
print(matrix)
```

```
accuracy = accuracy_score(test['Label'],predictions)
print(accuracy*100)
```

```
chart = classification_report(test['Label'],predictions)
```

```
print(chart)
```
